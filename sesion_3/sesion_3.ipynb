# Try-except (Try-catch)

```python
# Sin bloque try-except
edad_years_str= input("Digite su edad en años: ")
edad_years = int(edad_years_str)
edad_meses = edad_years * 12
print("Su edad en meses es: "+ str(edad_meses))
```

```python
# Con el bloque try-except
# Permite manejar errores sin que el programa se detenga
edad_years_str= input("Digite su edad en años: ")

try:
  edad_years = int(edad_years_str)
  edad_meses = edad_years * 12
  print("Su edad en meses es: "+ str(edad_meses))
except:
  print("Debe digitar un número")
```

# Assert

```python
# Es una instrucción que se utiliza para verificar si una condición se cumple.
# Es ideal cuando necesitamos asegurarnos de que los datos tengan un formato determinado o estén en un rango

peso = float(input("Digite su peso en kilogramos: "))
estatura = float(input("Digite su estatura en metros: "))

assert peso > 0, "El peso debe ser mayor a 0"
assert estatura > 0, "La estatura debe ser mayor a 0"

imc = peso / (estatura ** 2)
print("Su IMC es: "+ str(imc))
```

# List comprehensions

```python
# Crear una lista de los números del 1 al 10 elevados al cuadrado
# [1, 4, 9, 16, 25, 36, ...]

# [expresión for iterable if condición]

numeros_al_cuadrado = [x**2 for x in range(0,11) if x**2 % 2 == 0]

print(numeros_al_cuadrado)
```

### Operaciones matemáticas

```python
x = 26
y = -2

print("\n--- 1. ARITMÉTICA BÁSICA ---")
print("Potencia (np.power()):", np.power(x, 2))
print("Valor absoluto (np.abs()):", np.abs(-x))

print("\n--- 2. EXPONENCIALES Y LOGARITMOS ---")
print("Exponencial (np.exp()):", np.exp(x))
print("Logaritmo natural (np.log()):", np.log(x))
print("Logaritmo base 10 (np.log10()):", np.log10(x))
print("Logaritmo base 2 (np.log2()):", np.log2(x))

print("\n--- 3. REDONDEO Y DECIMALES ---")
print("Redondeo hacia abajo (np.floor()):", np.floor(x))
print("Redondeo hacia arriba (np.ceil()):", np.ceil(x))

print("\n--- 4. TRIGONOMETRÍA ---")
print("Seno (np.sin()):", np.sin(x))
print("Coseno (np.cos()):", np.cos(x))
print("Tangente (np.tan()):", np.tan(x))
print("Arcotangente (np.tan()):", np.atan(x))
print("Grados a radianes (np.deg2rad()):", np.deg2rad(180))
print("Radianes a grados (np.rad2deg()):", np.rad2deg(np.pi))
```

### Operaciones entre matrices

```python
# Transpuesta

print(matriz_ateatorios_ent)
print(matriz_ateatorios_ent.T)
```

```python
# Multiplicación @
matriz_ateatorios_1 = np.random.randint(1, 10, (3,3))
matriz_ateatorios_2 = np.random.randint(1, 10, (3,3))

print(matriz_ateatorios_1)
print(matriz_ateatorios_2)
print(matriz_ateatorios_1*matriz_ateatorios_2)
print(matriz_ateatorios_1@matriz_ateatorios_2)
```

```python
# Sacar la inversa de una matriz
print(np.linalg.inv(matriz_ateatorios_1))
```

```python
# Sistema de ecuaciones
# CASO: Resolver un sistema de ecuaciones
# 2x + 3y = 8
#  x -  y = 1

# Matriz A
A = np.array([[2,3],
              [1,-1]])

# Vector B
B = np.array([8,1])

# Solución
x = np.linalg.solve(A, B)
print(x)
```

## Operaciones de estadística

```python
# Operaciones estadísticas
datos_estatura= np.random.randint(150, 190, (800,1))
```

```python
# Media
print(np.mean(datos_estatura))

# Mediana
print(np.median(datos_estatura))

# Desviación estándar
print(np.std(datos_estatura))

# Varianza
print(np.var(datos_estatura))

# Valor mínimo y máximo
print(np.min(datos_estatura))
print(np.max(datos_estatura))

# Percentiles
print(np.percentile(datos_estatura, 80))
```

# Scipy

```python
from scipy.stats import skew, kurtosis

# Curtosis
print(kurtosis(datos_estatura))

# Asimetría
print(skew(datos_estatura))
```

```python
# Resolver integrales
from scipy import integrate

def funcion_integral(x):
  return np.log(x)

area_bajo_funcion = integrate.quad(funcion_integral, 0, 70 )
print(area_bajo_funcion)
```

```python
# Optimizar una función
def funcion_optimizar(x):
  return x**2 + 4*x - 9

from scipy import optimize

minimo = optimize.minimize(funcion_optimizar, x0=0)
print(minimo)
```
