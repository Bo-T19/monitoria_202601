## Agrupaciones

```python
precip = pd.read_excel('Precipitaci√≥n_20260223 (1).xlsx')
```

```python
precip.head(10)
```

```python
precip.info()
```

```python
precip= precip[['FechaObservacion','ValorObservado', 'Departamento', 'Municipio', 'Latitud', 'Longitud' ]]
precip.head(10)
```

```python
# Agrupar por municipio mostrando la media
media_precipitaciones= precip.groupby(['Departamento','Municipio']).mean(['ValorObservado'])
media_precipitaciones
```

```python
# Agrupar por municipio mostrando la media
media_precipitaciones= precip.groupby(['Departamento','Municipio']).sum(['ValorObservado'])
media_precipitaciones
```

```python
# Agrupar por fecha
precip['dia']= pd.DatetimeIndex(precip['FechaObservacion']).day
suma_precipitaciones= precip.groupby(['dia']).sum(['ValorObservado'])
suma_precipitaciones.drop(columns=['Latitud', 'Longitud'], inplace=True)
suma_precipitaciones
```

```python
# Borrar una columna de un dataframe
precip.drop(columns=['mes'], inplace=True)
```

# Folium üó∫

```python
import folium
```

```python
suma_precipitaciones= precip.groupby(['Municipio']).sum(['ValorObservado'])
suma_precipitaciones.reset_index(inplace=True)
suma_precipitaciones.drop(columns=['Latitud', 'Longitud', 'dia'], inplace=True)
```

```python
df_coordenadas = precip[['Municipio', 'Latitud', 'Longitud']]
```

```python
df_coordenadas= df_coordenadas.drop_duplicates(subset=['Municipio'])
```

```python
suma_precipitaciones= suma_precipitaciones.merge(df_coordenadas, left_on='Municipio', right_on='Municipio', how='left')
```

```python
suma_precipitaciones.info()
```

```python
# Calcular las coordenadas del centro geom√©trico del departamento
centro_lat = suma_precipitaciones['Latitud'].mean()
centro_lon = suma_precipitaciones['Longitud'].mean()

print(centro_lat, centro_lon)
```

```python
mapa_lluvia= folium.Map(location=[centro_lat, centro_lon], zoom_start=8)
```

```python
for index, row in suma_precipitaciones.iterrows():
  tamano_punto= row['ValorObservado']
  folium.CircleMarker(
    location=[row['Latitud'], row['Longitud']],
    radius=tamano_punto,
    color='blue',
    fill=True,
    fill_color='cyan',
    tooltip=f"{row['Municipio']}: {row['ValorObservado']}"
  ).add_to(mapa_lluvia)
```
