## Condicionales

```python
# Quiero hacer un programa para definir si un número es par o impar

numero_base=8

if numero_base % 2 == 0:
    print('Este es un número par')
else:
    print('Este es un número impar')
```

```python
# If - elif - else

palabra = 'Zapato'

if palabra[0] == 'A':
    print('Esta palabra empieza por A')
elif palabra[0] == 'B':
    print('Esta palabra empieza por B')
elif palabra[0] == 'C':
    print('Esta palabra empieza por C')
else:
    print('Esta palabra no empieza por A, B o C')
```

# Ejemplo con todo lo anterior

```python
# Datos de ventas
ventas = [
    {"tienda": "Tienda A", "producto": "Laptop", "cantidad": 5, "precio_unitario": 800},
    {"tienda": "Tienda B", "producto": "Tablet", "cantidad": 10, "precio_unitario": 300},
    {"tienda": "Tienda C", "producto": "Smartphone", "cantidad": 15, "precio_unitario": 600},
    {"tienda": "Tienda D", "producto": "Laptop", "cantidad": 3, "precio_unitario": 800},
    {"tienda": "Tienda E", "producto": "Smartphone", "cantidad": 8, "precio_unitario": 600}
]

# Calcular el $ total de las ventas
total_ventas=0
for item in ventas:
    total_ventas+=item['cantidad']*item['precio_unitario']
    print("Acumulado hasta el momento: "+str(total_ventas))

print(total_ventas)
```

```python
# Determinar la tienda que vendió más
max_ventas=0
tienda_ganadora=''

for item in ventas:
    ventas_actuales=item['cantidad']*item['precio_unitario']
    print('El máximo de ventas registrado hasta el momento es:' + str(max_ventas) + ', las ventas de la actual tienda fueron: ' + str(ventas_actuales))
    if ventas_actuales>max_ventas:
        max_ventas=ventas_actuales
        tienda_ganadora= item['tienda']
        print('La tienda '+ tienda_ganadora + ' ha establecido un nuevo record de ventas')

print ('La tienda con más ventas fue '+ tienda_ganadora)
```

# Funciones
Bloque de código reutilizable para realizar una tarea en específico

```python
# Definición de una función básica
def suma(a,b):
    """
    Función para realizar una suma de dos números
    """
    return a+b
```

```python
a= suma(5,6)
print(a)
```

```python
# Función más compleja
def calcular_promedio(nombre, notas, bono=0.1):

    promedio = sum(notas)/len(notas) 
    print ('El promedio de ' + nombre + " es: "+ str(promedio)+', con el bono es: ' +str(promedio + bono))
    return promedio + bono
```

```python
calcular_promedio('Santiago', [3, 4, 5, 3.5, 4])
```

```python
calcular_promedio('Santiago', [3, 4, 5, 3.5, 4], 0.5 )
```
