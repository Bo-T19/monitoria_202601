## Tipos de datos

```python
# Primer tipo: Enteros
x = 8
print (x, type(x))
```

```python
# Segundo tipo: Float
y = 10.789453465
print(y, type(y))
```

```python
# Suma de enteros y flotantes
print(x + y)
```

```python
# Tercer tipo: String
cadena_1 = "¡Hola mundo!"
cadena_2 = '!Hola mundo, otra vez!'
cadena_3 = """
¡Hola mundo!,
¿Cómo estan?
Yo muy bien
"""
```

```python
# Imprimir strings
print(cadena_1)
```

```python
print(cadena_2)
```

```python
print(cadena_3)
```

```python
# Cuarto tipo: Boooleanos
hoy_es_martes=True
hoy_es_miercoles=False

print(hoy_es_martes, type(hoy_es_martes))
```

```python
# Operaciones lógicas para generar booleanos
desigualdad = 5 > 9
print(desigualdad)
```

## Tipos de operaciones

```python
# Suma entre dos números
x + y
```

```python
# Suma entre dos strings
cadena_1 + cadena_2
```
